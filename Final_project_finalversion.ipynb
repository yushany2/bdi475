# Do some cleaning first

```python
modifieddf = df.fillna(' ') #I choose to replace any empty values with a space,and named the new dataset modifieddf
modifieddf.isnull().sum()# Make sure I did the last step successfully
modifieddf.to_csv('modifieddf.csv',index=False)# Store as a new csv
modifieddf.shape
modifieddf.info
```

#Filtering, sorting and aggregating


```python
modifieddf.groupby('Type 1').agg({'Attack': 'mean'})
modifieddf.sort_values('Speed', ascending=False, inplace=True)
modifieddf.head()
```

To see which type in type 1 has the highest/lowest Attack. Dragon is the highest.

```python
df_type1_hp = df.groupby('Type 1', as_index=False).agg({'HP': ['min', 'max', 'mean', 'sum', 'count', 'std']})

display(df_type1_hp)
```

I think  it is very useful for analyzing the overall level of each type. To see if there is a huge disparity in attack in each group from std, min and max.



# Type 1 Breakdown pie chart


```python
modifieddf.head(3)
```

```python
fig = px.pie(
    modifieddf[modifieddf['Type 1'] != ' '],
    names='Type 1',
    title='Type 1 Breakdown',
    width=600,
    height=600
)

fig.show()
```

After calculation of percentage of each kind of pokemonsin type 1, I know that Water is the most in type 1 and flying is the least.

## Type 1 Contribution


```python
fig = px.histogram(modifieddf, x='Type 1', title='Type1 Distribution', nbins=36)
fig.show()
```

The shape of Type 1 distribution does not follow a bell-curve distribution. I think developers randomly assign the number of pokemon s in each type in type 1.

## HP distribution

```python
fig = px.histogram(modifieddf, x='HP', title='HP Distribution', nbins=36)
fig.show()
```

Most of Pokemon have HP from 60-69.Just like most people are ordinary people.

```python
fig = px.scatter_3d(
    modifieddf,
    title='Sp. Atk,Sp. Def, Name 3D Scatter Plot',
    x="Sp. Atk",
    y="Sp. Def",
    z="Name",
    color="Name",
    template='plotly_dark',
    width=800,
    height=800
)
fig.show()
```

It is a colorful and complex 3D graph. It is interesting really to find a pokemon's name and see what Sp. Atk and Sp. Def they have.

```python
fig = px.density_heatmap(
    modifieddf,
    x='Attack',
    y='Speed',
    title='Attack vs Speed',
    height=600
)
fig.show()
```

The heatmap shows that most aggreate data for pokemons' speed and attach is Attack 60-79 and Speed 40-59 and there are totally 53 pokemons in this range.

```python
fig = px.box(modifieddf, y='Sp. Atk', title='Sp. Atk Dispersion (Vertical)')
fig.show()
```

Sp. Atk 194 is an outlier, median is 65 and the mins is 10.
